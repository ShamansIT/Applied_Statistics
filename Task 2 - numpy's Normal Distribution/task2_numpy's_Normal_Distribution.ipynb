# Numpy's Normal Distribution
***
### Description
This task checks whether the **numpy.random.normal()** function correctly generates values ​​from a normal distribution.

```python
# Import Libs
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Define distribution parameters

```python
mu = 10.0 # mean
sigma = 3.0 # standard deviation
n = 100000 # sample size

# Generate sample from the normal distribution
sample = np.random.normal(mu, sigma, n)
# Show
sample
```

## Shapiro-Wilk Test for Normality
The Shapiro-Wilk test checks the null hypothesis H<sub>0</sub>: the data comes from a normal distribution.

- If p-value > 0.05, then do not reject H<sub>0</sub>.
- If p-value ≤ 0.05, then reject H<sub>0</sub>.

```python
# Using a subsample of 5000 elements due to test limitations
subsample = sample[:5000]

# Performing the Shapiro-Wilk test
statistic, p_value = stats.shapiro(subsample)

print(f"Test statistic: {statistic}")
print(f"P-value: {p_value}")
```

## Interpretation of Test Results

```python
alpha = 0.05  # Significance level

if p_value > alpha:
    print("Cannot reject the null hypothesis: data is normally distributed.")
else:
    print("Reject the null hypothesis: data is not normally distributed.")
```

### Conclusion:
Based on the p-value obtained, can conclude the normality of the sample.

## Plotting the Histogram and Overlaying the Probability Density Function

```python
# Plotting the histogram with normalization
count, bins, ignored = plt.hist(sample, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Calculating values for the normal distribution's probability density function
normal_distrib = stats.norm.pdf(bins, mu, sigma)

# Overlaying the probability density function plot
plt.plot(bins, normal_distrib, 'r--', linewidth=2, label='Normal distribution')

# Adding labels and legend
plt.title('Sample Histogram and Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Displaying the plot
plt.show()
```

### Explanation:
The histogram shows the empirical distribution of the generated data.
The probability density function of the normal distribution is overlaid for comparison with the theoretical distribution.

## Overall Conclusion

The Shapiro-Wilk test and graphical analysis indicate that numpy.random.normal() correctly generates values from a normal distribution with the specified parameters μ = 10.0 and σ = 3.0.
